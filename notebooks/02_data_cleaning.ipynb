# Part 2: Data Cleaning & Preprocessing

## Objectives
1. Handle missing values through imputation
2. Convert data types (string to numeric)
3. Remove duplicates
4. Standardize date formats
5. Verify data integrity

```python
import pandas as pd
import numpy as np
import sys
sys.path.append('..')

from src.data_loader import load_csv, display_basic_info, save_csv
from src.preprocessing import convert_columns_to_numeric, impute_missing_values, standardize_date_column

df = load_csv('../data/raw/merged_file_FINAL.csv')
display_basic_info(df)
```

## Handle Missing Values & Convert Types

```python
# Check missing values
print("Missing values:")
print(df.isnull().sum())

# Impute SILVER_PRICE
imputation_indices = [1486, 7733, 7993, 8336]
df = impute_missing_values(df, 'SILVER_PRICE', imputation_indices)

# Convert to numeric
numeric_cols = ['GOLD_PRICE', 'GOLD_OPEN', 'GOLD_HIGH', 'GOLD_LOW',
                'SILVER_PRICE', 'VIX', 'GPR_INDEX', 'CPI_INFLATION']
df = convert_columns_to_numeric(df, numeric_cols)

# Standardize dates
df = standardize_date_column(df, 'DATE', dayfirst=True)

print(f"\nCleaned shape: {df.shape}")
print("\nRemaining nulls:")
print(df.isnull().sum().sum())
```

## Save Cleaned Data

```python
save_csv(df, '../data/processed/merged_final_cleaned.csv')
print("✅ Data cleaning complete!")
```
