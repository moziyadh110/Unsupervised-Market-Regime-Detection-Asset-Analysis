# Part 3: Exploratory Data Analysis (EDA)

## Objectives
1. Analyze distributions and summary statistics
2. Identify skewed variables
3. Examine correlations
4. Visualize time series patterns
5. Detect anomalies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append('..')

from src.data_loader import load_csv
from src.visualization import setup_plotting_style, plot_correlation_heatmap, plot_distribution, plot_time_series

setup_plotting_style()
df = load_csv('../data/processed/merged_final_cleaned.csv')
df['DATE'] = pd.to_datetime(df['DATE'])
```

## Summary Statistics

```python
print("Summary Statistics:")
print(df.describe())

print("\nSkewness:")
numeric_df = df.select_dtypes(include=[np.number])
print(numeric_df.skew().sort_values(ascending=False).head(10))
```

## Correlation Analysis

```python
# Correlation heatmap
plot_correlation_heatmap(df, figsize=(12, 10))

# Key correlations
corr = numeric_df.corr()
print("\nGold-related correlations:")
if 'GOLD_PRICE' in corr.columns:
    print(corr['GOLD_PRICE'].sort_values(ascending=False))
```

## Distributions of Key Variables

```python
key_vars = ['GOLD_PRICE', 'SILVER_PRICE', 'VIX', 'GPR_INDEX']
for var in key_vars:
    if var in df.columns:
        plot_distribution(df, var, bins=50)
```

## Time Series Visualization

```python
y_cols = ['GOLD_PRICE', 'SILVER_PRICE', 'VIX']
plot_time_series(df, 'DATE', y_cols, figsize=(14, 8))

print("✅ EDA complete!")
```
